# Clean the file to make it ready for dbt

```python
import pandas as pd

df = pd.read_csv("../data/csv_data_livsmedelsverket.csv")

print(df.head())
```

```python
df_copied = df.copy()
print(df_copied)
```

```python
df_melted = df_copied.melt(id_vars=["Livsmedelsnamn", "Livsmedelsnummer", "Gruppering"],
                           var_name="Näringsämne",
                           value_name="Värde")
```

```python
df_melted.to_csv("../data/csv_data_livsmedelsverket_melted.csv")
```
