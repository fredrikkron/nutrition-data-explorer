```python
import duckdb

df_gruppering_nulls = duckdb.query(
    """--sql
    SELECT
        Livsmedelsnamn,
        Gruppering
    FROM
        df
    WHERE Gruppering IS NULL
    GROUP BY Livsmedelsnamn, Gruppering
    """
)
df_gruppering_nulls # Gruppera till Övrigt i rensningen
```

```python
df_vitamin_null = duckdb.query(
    """--sql
    SELECT
        *
    FROM
        df
    WHERE Livsmedelsnamn = 'Hallonsylt lättsockrad'
    """
)

df_vitamin_null
```

```python
df_vitamin_null_2 = duckdb.query(
    """--sql
    SELECT
        *
    FROM
        df
    WHERE Livsmedelsnamn = 'Abborre filé panerad stekt'
    """
)

df_vitamin_null_2
```

```python
# Topp 10 Gruppering

top_10_category = duckdb.query(
    """--sql
    SELECT
        Gruppering,
        COUNT(*) AS antal
    FROM
        df
    GROUP BY Gruppering
    ORDER BY antal DESC 
    LIMIT 10
    """
)

top_10_category
```
